```python
import pandas as pd

# Load từng file .txt vào DataFrame riêng biệt
customer_202311_df = pd.read_csv('D:\\HKT\\HKT_Architect\\hackathon data\\customer\\customer_202311.txt', delimiter='|')
customer_202312_df = pd.read_csv('D:\\HKT\\HKT_Architect\\hackathon data\\customer\\customer_202312.txt', delimiter='|')
customer_202401_df = pd.read_csv('D:\\HKT\\HKT_Architect\\hackathon data\\customer\\customer_202401.txt', delimiter='|')
customer_202402_df = pd.read_csv('D:\\HKT\\HKT_Architect\\hackathon data\\customer\\customer_202402.txt', delimiter='|')

city_df = pd.read_csv('D:\\HKT\\HKT_Architect\\hackathon data\\master\\city.txt', delimiter='|')
product_df = pd.read_csv('D:\\HKT\\HKT_Architect\\hackathon data\\master\\product.txt', delimiter='|')

transaction_202311_df = pd.read_csv('D:\\HKT\\HKT_Architect\\hackathon data\\transaction\\transaction_202311.txt', delimiter='|')
transaction_202312_df = pd.read_csv('D:\\HKT\\HKT_Architect\\hackathon data\\transaction\\transaction_202312.txt', delimiter='|')
transaction_202401_df = pd.read_csv('D:\\HKT\\HKT_Architect\\hackathon data\\transaction\\transaction_202401.txt', delimiter='|')
```

```python
import pandasql as ps

# Kiểm tra cột 'Customer_ID' trong các DataFrame của customer
for name, df in {"customer_202311": customer_202311_df, "customer_202312": customer_202312_df, 
                 "customer_202401": customer_202401_df, "customer_202402": customer_202402_df}.items():
    query = """
    SELECT 
        COUNT(Customer_ID) AS total_count,
        COUNT(DISTINCT Customer_ID) AS distinct_count

    FROM df
    """
    result = ps.sqldf(query, {"df": df})
    print(f"Kết quả cho {name}:")
    print(result)
    print("\n")

# Kiểm tra cột 'City_ID' trong bảng city
query = """
SELECT 
    COUNT(CityId) AS total_count,
    COUNT(DISTINCT CityId) AS distinct_count
FROM city_df
"""
result = ps.sqldf(query, {"city_df": city_df})
print("Kết quả cho city:")
print(result)
print("\n")

# Kiểm tra cột 'Product_ID' trong bảng product
query = """
SELECT 
    COUNT(Product_ID) AS total_count,
    COUNT(DISTINCT Product_ID) AS distinct_count
FROM product_df
"""
result = ps.sqldf(query, {"product_df": product_df})
print("Kết quả cho product:")
print(result)
print("\n")

# Kiểm tra cột 'Transaction_ID' trong các DataFrames của transaction
for name, df in {"transaction_202311": transaction_202311_df, "transaction_202312": transaction_202312_df, 
                 "transaction_202401": transaction_202401_df}.items():
    query = """
    SELECT 
        COUNT(Transaction_ID) AS total_count,
        COUNT(DISTINCT Transaction_ID) AS distinct_count
    FROM df
    """
    result = ps.sqldf(query, {"df": df})
    print(f"Kết quả cho {name}:")
    print(result)
    print("\n")
```

```python
for name, df in {"customer_202311": customer_202311_df, "customer_202312": customer_202312_df, 
                 "customer_202401": customer_202401_df, "customer_202402": customer_202402_df}.items():
    query = """
    SELECT 
        COUNT(*) AS total_rows,
        SUM(CASE WHEN Age < 0 OR Age > 120 THEN 1 ELSE 0 END) AS out_of_range_count
    FROM df
    """
    result = ps.sqldf(query, {"df": df})
    total_rows = result['total_rows'][0]
    out_of_range_count = result['out_of_range_count'][0]
    
    print(f"Kết quả cho DataFrame '{name}':")
    print(f"  - Tổng số hàng: {total_rows}")
    print(f"  - Số lượng giá trị Age ngoài khoảng 0-120: {out_of_range_count}")
    print("\n")
```

```python
import pandasql as ps

# Danh sách các cột cần kiểm tra NULL
columns_to_check = ["Customer_ID", "Name", "Email", "Phone", "Address", "City_ID", "Age", "Gender", "Income", "Customer_Segment"]

# Kiểm tra số lượng giá trị NULL trong từng cột cho mỗi DataFrame customer
for name, df in {"customer_202311": customer_202311_df, "customer_202312": customer_202312_df, 
                 "customer_202401": customer_202401_df, "customer_202402": customer_202402_df}.items():
    
    # Dùng vòng lặp để kiểm tra từng cột
    for col in columns_to_check:
        query = f"""
        SELECT 
            COUNT(*) AS total_rows,
            SUM(CASE WHEN {col} IS NULL THEN 1 ELSE 0 END) AS null_count
        FROM df
        """
        result = ps.sqldf(query, {"df": df})
        null_count = result['null_count'][0]
        
        print(f"  '{col}': {null_count} NULL")
    
    print("\n")
```
